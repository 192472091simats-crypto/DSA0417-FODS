```python
# Lists of item prices and quantities
prices = [10.50, 5.25, 3.75]      # example prices
quantities = [2, 4, 3]           # example quantities

# Discount and tax rates (in percentage)
discount_rate = 10    # 10% discount
tax_rate = 8          # 8% tax

# Step 1: Calculate subtotal
subtotal = 0
for i in range(len(prices)):
    subtotal += prices[i] * quantities[i]

# Step 2: Calculate discount
discount_amount = subtotal * (discount_rate / 100)

# Step 3: Subtotal after discount
price_after_discount = subtotal - discount_amount

# Step 4: Calculate tax
tax_amount = price_after_discount * (tax_rate / 100)

# Step 5: Final total cost
total_cost = price_after_discount + tax_amount

# Display results
print("Subtotal:", round(subtotal, 2))
print("Discount Amount:", round(discount_amount, 2))
print("Tax Amount:", round(tax_amount, 2))
print("Total Cost:", round(total_cost, 2))
```
